# 1.2 Computational Neuroscience: Descriptive Models

**Definition of Computational Neuroscience:**
- **Terry Sejnowski's Definition:** Explains in computational terms how brains generate behaviors.
- **Peter Dayan and Larry Abbott's Definition:** 
  - Characterizes what nervous systems do.
  - Determines how they function.
  - Understands why they operate in particular ways.
  
  These aspects align with descriptive, mechanistic, and interpretive models discussed earlier.

**Concept of Receptive Fields:**

1. **Historical Experiment by Hubel and Wiesel:**
   - **Objective:** Study the visual system of cats.
   - **Method:** Implanted electrodes in the visual cortex of cats to record neural responses.
   - **Findings:** Specific neurons respond robustly to certain stimuli, like a bar of light at a specific angle.

2. **Receptive Field Definition:**
   - The specific properties of a sensory stimulus that cause a strong response from a neuron.
   - **Examples:**
     - Retina: Cells respond to light spots at certain locations.
     - Visual Cortex: Cells respond to bars of light oriented at specific angles.

3. **Descriptive Models of Receptive Fields:**
   - **Retina Example:** 
     - **Center-Surround Receptive Fields:**
       - **On-Center, Off-Surround:** Neurons respond strongly when light is on the center or off in the surrounding area.
       - **Off-Center, On-Surround:** Neurons respond strongly when light is off in the center or on in the surrounding area.
   - **Primary Visual Cortex Example:**
     - **Oriented Receptive Fields:** Neurons respond best to bars of light oriented at specific angles (e.g., 45 degrees).

4. **Orientation Variation in the Visual Cortex:**
   - Receptive fields vary in orientation (vertical, horizontal, various angles).
   - Some cells prefer dark bars, while others prefer bright bars.

5. **Future Topic: Mechanistic Models**
   - **Objective:** Understand how center-surround receptive fields in the retina and LGN lead to oriented receptive fields in the visual cortex.

**Summary:**
The lecture introduces computational neuroscience, focusing on receptive fields as a key concept. Receptive fields describe how specific stimuli generate robust responses in neurons. Descriptive models are used to understand these fields, as demonstrated by Hubel and Wiesel's experiments on the retina and visual cortex. Mechanistic models, to be covered next, will explore how these receptive fields develop.

# 1.3 Computational Neuroscience: Mechanistic and Interpretive Models

**Recap of Previous Lecture:**
- **Descriptive Models:** Center-surround and oriented receptive fields.
- **Question:** How are oriented receptive fields constructed from center-surround receptive fields?

**Mechanistic Model:**

1. **Neuroanatomy of Visual System:**
   - **Pathway:** Visual information flows from the retina through the optic nerve to the lateral geniculate nucleus (LGN) and then to the primary visual cortex (V1).
   - **Receptive Fields:**
     - **LGN:** Center-surround receptive fields.
     - **V1:** Elongated, oriented receptive fields.

2. **Construction of Oriented Receptive Fields:**
   - **Convergence:** Multiple LGN cells converge onto a single V1 cell.
   - **Mechanism:** The arrangement of center-surround receptive fields from LGN cells creates an elongated, oriented receptive field in V1.
   - **Example:** A V1 cell responds to bars of light at specific orientations due to the arrangement of its input from LGN cells.

3. **Limitations:**
   - The Hubel and Wiesel model focuses on feedforward connections but does not account for recurrent connections within V1.
   - Recurrent inputs also contribute to the responses of V1 cells and will be discussed later.

**Interpretive Model:**

1. **Efficient Coding Hypothesis:**
   - **Goal:** Represent images as faithfully and efficiently as possible using available neurons.
   - **Optimization:** Find receptive fields that minimize reconstruction errors of natural images and ensure response independence.

2. **Efficient Coding Algorithms:**
   - **Sparse Coding (Olshausen and Field)**
   - **Independent Component Analysis (Bell and Sejnowski)**
   - **Predictive Coding (Sejnowski and Ballard)**

3. **Results:**
   - **Receptive Fields:** Receptive fields optimized by efficient coding algorithms resemble those observed in V1, being oriented and spatially localized.
   - **Conclusion:** The brainâ€™s receptive fields are tuned to efficiently represent natural images, which might also apply to auditory cortex representations.

# 1.4 The Electrical Personality of Neurons

In this Neurobiology 101 lecture, several key concepts about neurons and their functions were covered:

1. **Neuron Structure**:
   - Neurons are the primary cells in the brain, acting as the main units for communication.
   - **Dendrites**: Branch-like structures that receive signals from other neurons.
   - **Axon**: A slender extension that transmits signals away from the neuron.
   - **Cell Body**: Contains the nucleus and integrates signals received by the dendrites.

2. **Neuron Types**:
   - **Pyramidal Neurons**: Found in the visual cortex; named for their triangular cell bodies.
   - **Purkinje Cells**: Located in the cerebellum; characterized by their extensive branching dendrites.
   - **Optic Tectum Cells**: Have diverse branching structures depending on their location.

3. **Neuron Doctrine**:
   - States that neurons are discrete units of the brain and not continuous with other cells.
   - Information flows from dendrites through the cell body and along the axon.

4. **Electrical Properties**:
   - Neurons maintain a resting membrane potential of approximately -70 mV.
   - This potential difference is due to ionic gradients maintained by pumps and channels in the cell membrane.

5. **Ionic Channels**:
   - **Voltage-Gated Channels**: Open or close in response to changes in membrane voltage.
   - **Chemically-Gated Channels**: Open or close in response to the binding of chemicals.
   - **Mechanically-Gated Channels**: Open or close in response to mechanical pressure or stretch.

6. **Action Potential**:
   - Generated when a neuron reaches a threshold level of depolarization.
   - Involves a rapid influx of sodium ions followed by an outflux of potassium ions.
   - Propagates along the axon through a process called saltatory conduction, facilitated by myelin sheaths and Nodes of Ranvier.

7. **Myelination**:
   - **Myelin Sheaths**: Insulating layers around axons that speed up signal transmission.
   - **Nodes of Ranvier**: Gaps in the myelin sheath where action potentials are regenerated.
   - **Saltatory Conduction**: The process of action potentials jumping from one Node of Ranvier to the next, enhancing speed and efficiency.

8. **Disease Example**:
   - **Multiple Sclerosis**: A condition where myelin is damaged, impairing signal transmission and causing various symptoms.

# 1.5 Making Connections Synapses

1. **Ion Channels and Action Potentials**:
   - **Action potentials (spikes)** are generated by ion channels, specifically sodium (Na+) and potassium (K+) channels. 
   - These action potentials propagate along the axon of a neuron, transmitting electrical signals.

2. **Synapses**:
   - **Electrical Synapses**: 
     - Function via **gap junctions**, which are direct ionic channels connecting two neurons.
     - Enable fast transmission and synchronization between neurons (useful in escape reflexes).
   - **Chemical Synapses**:
     - Use **neurotransmitters** (e.g., glutamate) stored in vesicles.
     - When an action potential reaches the presynaptic neuron, vesicles release neurotransmitters into the **synaptic cleft**.
     - Neurotransmitters bind to receptors on the postsynaptic neuron, causing ion channels to open and alter membrane potential.
     - Chemical synapses allow modulation, making them essential for learning and memory.

3. **Excitatory and Inhibitory Synapses**:
   - **Excitatory Synapses**: Increase the postsynaptic membrane potential, promoting neuron firing.
   - **Inhibitory Synapses**: Decrease the postsynaptic membrane potential, reducing neuron firing.

4. **Hebbian Plasticity (Synaptic Plasticity)**:
   - Proposed by Donald Hebb: "Neurons that fire together wire together."
   - **Hebbian Rule**: If neuron A repeatedly contributes to the firing of neuron B, the synapse from A to B is strengthened.
   - Example: Predictive associations, such as hearing a growl and anticipating danger.
   - **Long-Term Potentiation (LTP)**:
     - An increase in synaptic strength lasting for hours/days.
     - Experimentally observed as an increase in EPSP (Excitatory Postsynaptic Potential) size after repeated stimulation.
   - **Long-Term Depression (LTD)**:
     - A decrease in synaptic strength, lasting for hours/days.
     - Observed as a reduction in EPSP size with repeated stimulation that weakens the connection.

5. **Spike-Timing Dependent Plasticity (STDP)**:
   - **Timing of spikes** between neurons is crucial for determining whether LTP or LTD occurs.
   - **LTP**: When the input (neuron A) fires before the output (neuron B).
   - **LTD**: When the input fires after the output.
   - **Timing window**: The critical window for STDP is around 40 milliseconds.

6. **Synapses and Memory**:
   - Synapses are believed to be the basis for memory and learning, as proposed by the **Synapse Doctrine**.
   - Experiences and skills are stored as changes in synaptic strength in neural networks.

7. **Current Research**:
   - Ongoing large-scale brain projects (e.g., Human Brain Project, Brain Initiative) aim to map and simulate brain activity to understand how neural networks underlie perception, behavior, and consciousness.

# 1.6 Time to Network: Brain Areas and their Function

1. **Peripheral Nervous System (PNS)**:
   - Divided into two main components: **somatic** and **autonomic nervous systems**.
   - **Somatic Nervous System**:
     - Involves voluntary muscle control and sensory information.
     - Nerves are bundles of axons, transmitting signals between the CNS and peripheral muscles.
     - Example: Shaking a friend's hand involves afferent nerve fibers (carrying sensory information to the CNS) and efferent nerve fibers (carrying motor commands from the CNS to the muscles).
   - **Autonomic Nervous System**:
     - Controls involuntary functions, such as heart rate, digestion, and respiratory rate.
     - Operates below the level of consciousness and includes the "fight or flight" response.

2. **Central Nervous System (CNS)**:
   - Composed of the **spinal cord** and the **brain**.
   - **Spinal Cord**:
     - Responsible for reflex arcs, such as withdrawing your hand from a hot object (reflexive action occurs via the spinal cord before conscious awareness).
     - Acts as a conduit for motor commands from the brain and sensory feedback from the body.
   - **Brain**:
     - **Hindbrain**:
       - **Medulla Oblongata**: Controls vital functions like breathing and blood pressure.
       - **Pons**: Involved in sleep, arousal, and connection to the cerebellum.
       - **Cerebellum**: Coordinates voluntary movement, balance, and cognitive functions like language.
     - **Midbrain**:
       - Controls eye movements, auditory and visual reflexes, and basic functions like sleep and arousal.
     - **Thalamus**:
       - Relay station for sensory information (except smell) and regulates sleep and wakefulness.
     - **Hypothalamus**:
       - Regulates basic survival functions: fighting, fleeing, feeding, and mating (the "four F's").
     - **Cerebrum**:
       - Includes the **cerebral cortex**, **basal ganglia**, **hippocampus**, and **amygdala**.
       - Involved in perception, action, learning, memory, and higher cognitive functions.
       - **Cerebral Cortex**:
         - A layered structure with about 30 billion neurons and 300 trillion connections.
         - Organized in six layers with specific input/output patterns, suggesting a common computational principle across different cortical areas.

3. **Neural vs. Digital Computing**:
   - **Device Count**: The brain has 100 billion neurons, each with up to 10,000 connections, surpassing digital circuits in connectivity.
   - **Device Speed**: Digital circuits (e.g., 100 picosecond clock speeds) are faster than biological neurons (100 microsecond resolution).
   - **Computing Paradigms**:
     - The brain uses massively parallel computation and adaptive connectivity through synapses.
     - Digital computers generally use sequential processing with fixed connectivity.
   - **Strengths**: 
     - Digital computers excel at math and symbolic processing.
     - The brain excels at solving ill-posed problems like perception and language processing.

4. **Conclusion**:
   - While much is still unknown about the computational basis of the brain, understanding its structure and function gives insight into its unique capabilities compared to digital systems.
   - Future research, including techniques like electrophysiology and functional imaging, may help unravel the complexities of brain computation.

This summary provides an overview of the peripheral and central nervous systems, their components, and how the brain's computational model compares to digital computing.
